```python
columns_to_analyze = df.columns[df.columns.str.startswith('RHI')]
columns_to_analyze
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(12, 8))
plt.scatter(columns_to_analyze, df[columns_to_analyze].mean(), marker='x')
plt.xlabel('Death Cause Name')
plt.ylabel('Mean of Death index')
plt.xticks(columns_to_analyze, rotation=90)
plt.grid(True)
plt.show()
```

### 🤔 looks like 'non hispanic' have low death rate index. 'Homicide is also low'. 
### Coronary Heart Disease (CHD) and Stroke have Very High death index than other

```python
# lets see about CHD and Stroke
```

```python
df.head()
```

```python
df.groupby('CHSI_State_Name')['RHI_CHD_Ind', 'RHI_Stroke_Ind'].agg(np.mean)
```

### lets visualize this in plot

```python
state_name = df['CHSI_State_Name'].unique()
state_name
```

```python
plt.figure(figsize=(12, 8))
plt.bar(state_name, df.groupby('CHSI_State_Name')['RHI_CHD_Ind'].agg(np.mean), color='r')
plt.plot(state_name, df.groupby('CHSI_State_Name')['RHI_Stroke_Ind'].agg(np.mean), color='b')
plt.xlabel('State Name')
plt.ylabel('Mean of death Index')
plt.xticks(state_name, rotation=90)
plt.show()
```

# looks similar in all the states

# lets see for cancers

```python
df.columns
```

```python
col_cancers = ['RHI_Brst_Cancer_Ind', 'RHI_Col_Cancer_Ind', 'RHI_Lung_Cancer_Ind']
df[col_cancers].mean()
```

# Lung cancer > colonal cancer > breast cancer

```python
# lets see death index of these cancers in diff states

df.groupby('CHSI_State_Name')[col_cancers].agg(np.mean)
```

# We can plot this to give eye some relief from these numbers 😁

```python
plt.figure(figsize=(12, 5))
plt.scatter(state_name, df.groupby('CHSI_State_Name')['RHI_Brst_Cancer_Ind'].agg(np.mean), marker='o')
plt.plot(state_name, df.groupby('CHSI_State_Name')['RHI_Col_Cancer_Ind'].agg(np.mean), color='b')
plt.scatter(state_name, df.groupby('CHSI_State_Name')['RHI_Lung_Cancer_Ind'].agg(np.mean), marker='x')
plt.xlabel('State Name')
plt.ylabel('Mean of death Index')
plt.xticks(state_name, rotation=90)
plt.legend(['RHI_Lung_Cancer_Ind', 'RHI_Col_Cancer_Ind', 'RHI_Brst_Cancer_Ind'])
plt.show()
```

## Alaska have very low breast cancer death index than other states``
