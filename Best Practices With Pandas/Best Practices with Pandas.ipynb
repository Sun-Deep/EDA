# Which talk provoke the most online discussion?

```python
ted.head()
```

```python
ted[ted.comments == ted.comments.max()]
```

```python
# from this -> row number 96 seems have most number of comments.
# lets normalize comments with respect to views so that we can decide how provoking is the talk to start discussion.
```

```python
ted['comments_per_view'] = ted.comments / ted.views
```

```python
ted.talks_per_view
```

```python
# least popular
ted.talks_per_view.sort_values().head(1)
```

```python
# most popular for discussion
ted.talks_per_view.max()
```

```python
ted[ted.talks_per_view == ted.talks_per_view.max()]
```

```python
# adding column view per comment
ted['views_per_comment'] = ted.views / ted.comments
```

```python
ted.views_per_comment
```

```python
ted.views_per_comment.sort_values().head()
```
